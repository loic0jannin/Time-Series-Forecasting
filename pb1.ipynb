```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data for AMZ
amzn_data = pd.read_csv('stock_data_AMEN.csv')
amzn_data['Date'] = pd.to_datetime(amzn_data['Date'])
amzn_data.set_index('Date', inplace=True)

# Load data for NCLH
nclh_data = pd.read_csv('stock_data.csv')
nclh_data['Date'] = pd.to_datetime(nclh_data['Date'])
nclh_data.set_index('Date', inplace=True)
```

```python
def plot_price_and_volume(stock_data, stock_name, year):
    # Filter data for the specified year
    stock_data_year = stock_data[stock_data.index.year == year]

    # Create subplots with two axes
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(12, 8))
    
    # Plotting close prices
    ax1.plot(stock_data_year.index, stock_data_year['Close'], label=f'{stock_name} Close', color='blue')
    ax1.set_ylabel('Close Price')
    
    # Plotting volumes
    ax2.plot(stock_data_year.index, stock_data_year['Volume'], label=f'{stock_name} Volume', color='red')
    ax2.set_ylabel('Volume')
    
    plt.suptitle(f'{stock_name} Close Prices and Volumes for {year}')
    plt.xlabel('Date')
    
    # Show the plot
    plt.show()
```

```python

plot_price_and_volume(amzn_data, 'AMZN', 2016)
plot_price_and_volume(amzn_data, 'AMZN', 2017)
plot_price_and_volume(amzn_data, 'AMZN', 2018)
plot_price_and_volume(amzn_data, 'AMZN', 2019)
plot_price_and_volume(amzn_data, 'AMZN', 2020)

plot_price_and_volume(nclh_data, 'NCLH', 2016)
plot_price_and_volume(nclh_data, 'NCLH', 2017)
plot_price_and_volume(nclh_data, 'NCLH', 2018)
plot_price_and_volume(nclh_data, 'NCLH', 2019)
plot_price_and_volume(nclh_data, 'NCLH', 2020)
```

```python
def plot_volumes_2018_to_2020(stock_data_amzn, stock_data_nclh):
    # Extract data for the specified years (2018 to end of 2020)
    amzn_data_2018_2020 = stock_data_amzn['2018-01-01':'2020-12-31']
    nclh_data_2018_2020 = stock_data_nclh['2018-01-01':'2020-12-31']

    # Plotting volumes
    plt.figure(figsize=(12, 6))

    plt.subplot(2, 1, 1)
    plt.plot(amzn_data_2018_2020.index, amzn_data_2018_2020['Volume'], label='AMZN', color='blue')
    plt.title('Volume for AMZN (2018-2020)')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend(loc='upper left')
    plt.grid(True)

    plt.subplot(2, 1, 2)
    plt.plot(nclh_data_2018_2020.index, nclh_data_2018_2020['Volume'], label='NCLH', color='red')
    plt.title('Volume for NCLH (2018-2020)')
    plt.xlabel('Date')
    plt.ylabel('Volume')
    plt.legend(loc='upper left')
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Assuming you have loaded the data for AMZN and NCLH, you can call the function like this:
plot_volumes_2018_to_2020(amzn_data, nclh_data)
```
